```python
## Task 10  Знайдіть повторювані записи (від 2-го випадку повторення і далі) у заданому масиві і позначте їх як True. Перше повторення має бути False.

# Input np.random.seed(100) a = np.random.randint(0, 5, 10)

## [0 0 3 0 2 4 2 2 2 2] #Output [False True False True False False True True True True]

np.random.seed(100)
a = np.random.randint(0, 5, 10)

seen = set()
result = []

for x in a:
    if x in seen:
        result.append(True)
    else:
        result.append(False)
        seen.add(x)

result = np.array(result)

print(a)
print(result)

```
